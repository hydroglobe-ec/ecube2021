## Purpose
This notebook demonstrates HydroGlobe, a Jupyter Notebook tool delivering ready-to-use earth observation data for hydrological modelers with a stand-alone easy to use web interface.

## Technical contributions

*   A Jupyter Notebook tool providing spatially-averaged time series of earth observation data eliminating the need for hydrological modelers to perform complex processing tasks manually
*   An example of making the notebook a stand-alone web application with intuitive user experience by enhancing and extending default Jupyter widgets


## Methodology
1. HydroGlobe takes the following inputs: (i) a watershed shapefile, (ii) an earth observation product type, and (iii) temporal extent in the form of start/end date.
2. Sinusoidal tile grid of MODIS tiling system is used to calculate the overlapped area with the input shapefile.
3. With the provided inputs and collected grids, HydroGlobe fetches remotely sensed earth observation data from the corresponding remote data sources, such as Numerical Terradynamic Simulation Group (NTSG), or United States Geological Survey (USGS).
4. Spatially averaged time series of earth observations are computed for each sub-basin.
5. HydroGlobe generates final outputs including csv files and interactive map visualization.

## Results
HydroGlobe generated two types of outputs: 1) CSV formatted geospatially aggregated time series data computed from earth observation data which can be downloaded, and 2) GeoJson data for heat map-based interactive visualization.

## Downloading ET Data for Cedar Creek Watershed in Indiana on Jan. 2010.
By completing this task, you will be able to download remotely sensed evapotranspiration (ET) data for any watershed. A watershed may have multiple sub-basins. Therefore, the GIS shapefile of the watershed may have multiple polygons, i.e., one polygon representing the boundary for each sub-basin. Taking such a GIS shapefile as input, HydroGlobe tool will download ET time series for each of the sub-basins within a watershed.

This task will download ET data for the Cedar Creek watershed in Indiana, USA. To get started, click here to download the [shapefile](https://d32ogoqmya1dw8.cloudfront.net/files/hydromodules/steps/shapefile.v2.zip). This shapefile has 51 polygons representing the 51 sub-basins in Cedar Creek watershed.

Open HydroGlobe tool via [Binder Link](https://mybinder.org/v2/gh/hydroglobe-ec/ecube2021/HEAD?urlpath=apps%2Fbin%2FHydroglobe.ipynb).
